### Exons distribution and psi

```python
# required packages
import pandas as pd
%matplotlib inline
```

```python
# data
control = pd.read_table('snakemake_res/exons_genes_psi_p_control.bed', header = None)
xrn1 = pd.read_table('exons_p_genes_xrn1.bed', header = None)
smg6xrn1 = pd.read_table('exons_p_genes_smg6xrn1.bed', header = None)
upf1xrn1 = pd.read_table('exons_p_genes_upf1xrn1.bed', header = None)
```

```python
control.head()
```

```python
xrn1.head()
```

```python
control_sub = control.iloc[:,[6,1,2,10,13,11]]
xrn1_sub = xrn1.iloc[:,[10,11]]
smg6xrn1_sub = smg6xrn1.iloc[:,[10,11]]
upf1xrn1_sub = upf1xrn1.iloc[:,[10,11]]
control_sub.columns = ['Gene', 'Gene_start', 'Gene_stop', 'Exon', 'p', 'control_psi']
xrn1_sub.columns = ['Exon', 'xrn1_psi']
smg6xrn1_sub.columns = ['Exon', 'smg6xrn1_psi']
upf1xrn1_sub.columns = ['Exon','upf1xrn1_psi']
```

```python
xrn1_sub.head()
```

#### XRN1 vs. control

```python
control_xrn1 = pd.merge(control_sub, xrn1_sub, on="Exon")
```

```python
control_xrn1['delta-psi'] = control_xrn1['xrn1_psi'] - control_xrn1['control_psi']
control_xrn1 = control_xrn1.dropna()
```

```python
control_xrn1.head()
```

```python
control_xrn1_non_zero = control_xrn1[control_xrn1['delta-psi'] != 0]

control_xrn1_non_zero.plot(x="p", y="delta-psi", kind='scatter', title = 'xrn1 vs. control (zero delta-psi are not shown)' )
```

#### SMG6XRN1 vs. control

```python
control_smg6xrn1 = pd.merge(control_sub, smg6xrn1_sub, on="Exon")
```

```python
control_smg6xrn1['delta-psi'] = control_smg6xrn1['smg6xrn1_psi'] - control_smg6xrn1['control_psi']
control_smg6xrn1 = control_smg6xrn1.dropna()
```

```python
control_smg6xrn1_non_zero = control_smg6xrn1[control_smg6xrn1['delta-psi'] != 0]

control_smg6xrn1_non_zero.plot(x="p", y="delta-psi", kind='scatter', title = 'smg6xrn1 vs. control (zero delta-psi are not shown)' )
```

#### UPF1XRN1 vs. control

```python
control_upf1xrn1 = pd.merge(control_sub, upf1xrn1_sub, on="Exon")
```

```python
control_upf1xrn1['delta-psi'] = control_upf1xrn1['upf1xrn1_psi'] - control_upf1xrn1['control_psi']
control_upf1xrn1 = control_upf1xrn1.dropna()
```

```python
control_upf1xrn1_non_zero = control_upf1xrn1[control_upf1xrn1['delta-psi'] != 0]

control_upf1xrn1_non_zero.plot(x="p", y="delta-psi", kind='scatter', title = 'upf1xrn1 vs. control (zero delta-psi are not shown)' )
```

#### SMG6XRN1 vs. XRN1

```python
xrn1_smg6xrn1 = pd.merge(xrn1_sub, smg6xrn1_sub, on="Exon")
```

```python
xrn1_smg6xrn1 = pd.merge(xrn1_smg6xrn1, control_sub, on="Exon")
```

```python
xrn1_smg6xrn1.drop(['Gene', 'Gene_start', 'Gene_stop', 'control_psi'], axis=1, inplace=True)
```

```python
xrn1_smg6xrn1['delta-psi'] = xrn1_smg6xrn1['smg6xrn1_psi'] - xrn1_smg6xrn1['xrn1_psi']
xrn1_smg6xrn1 = xrn1_smg6xrn1.dropna()
```

```python
xrn1_smg6xrn1_non_zero = xrn1_smg6xrn1[xrn1_smg6xrn1['delta-psi'] != 0]

xrn1_smg6xrn1_non_zero.plot(x="p", y="delta-psi", kind='scatter', title = 'smg6xrn1 vs. xrn1 (zero delta-psi are not shown)' )
```

### UPF1XRN1 vs. XRN1

```python
xrn1_upf1xrn1 = pd.merge(xrn1_sub, upf1xrn1_sub, on="Exon")
xrn1_upf1xrn1 = pd.merge(xrn1_upf1xrn1, control_sub, on="Exon")
xrn1_upf1xrn1.drop(['Gene', 'Gene_start', 'Gene_stop', 'control_psi'], axis=1, inplace=True)

```

```python
xrn1_upf1xrn1['delta-psi'] = xrn1_upf1xrn1['upf1xrn1_psi'] - xrn1_upf1xrn1['xrn1_psi']
xrn1_upf1xrn1 = xrn1_upf1xrn1.dropna()
```

```python
xrn1_upf1xrn1_non_zero = xrn1_upf1xrn1[xrn1_upf1xrn1['delta-psi'] != 0]

xrn1_upf1xrn1_non_zero.plot(x="p", y="delta-psi", kind='scatter', title = 'upf1xrn1 vs. xrn1 (zero delta-psi are not shown)' )
```
